### Since we have to plot graphs again and again it is better to make a function to plot graphs

```python
# making a function to plot the graphs
def plt_frequency(x_var,data_df,x_label,title,hue_var=None,rot_flag=0,fig_size_x=15,fig_size_y=6):
    ncount=len(data_df)
    plt.figure(figsize=(fig_size_x,fig_size_y))    
    ax=sns.countplot(x=x_var, data=data_df,hue=hue_var,palette="mako")
    plt.title(title)
    plt.xlabel(x_label)
    #for printing percentage over each bar
    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text
    if(rot_flag==1):
        ax.set_xticklabels(ax.get_xticklabels(), rotation=50, ha="right", fontsize=12)
    plt.show()
```

```python
def dti_categories(dti):
    if dti <= 10.00:
        return 'low'
    elif (dti >10.00 and dti <= 20.00):
        return 'medium'
    else:
        return 'high'
    
def experience_level(exp):
    if exp <= 3:
        return 'Junior'
    elif (exp >3 and exp <= 7):
        return 'Senior'
    else:
        return 'Specialist'
    
def int_rate_categories(intrest):
    if intrest <= 12:
        return 'low'
    elif (intrest >12 and intrest <= 18):
        return 'medium'
    else:
        return 'high'
    
loan['dti_cat'] = loan['dti'].apply(lambda x : dti_categories(x) )
loan['exp_level'] = loan['emp_length'].apply(lambda x : experience_level(int(x)))
loan['int_rate_seg'] = loan['int_rate'].apply(lambda x : int_rate_categories(int(x)))
```

```python
def create_subplot(x_var,data_df,x_label,title,hue_var=None,palette_var='mako',rot_flag=0,axis=None):
    ncount=len(data_df)
    if axis is None:
        axis= plt.gca()  
    bp = sns.countplot(x=x_var, data=data_df,hue=hue_var,palette="mako",ax=axis)
    if(rot_flag==1):
        axis.set_xticklabels(axis.get_xticklabels(), rotation=40, ha="right", fontsize=12)
        plt.tight_layout()
    axis.set_title(title)
    axis.set_xlabel(x_label)
    axis.set_ylabel('Frequency [%]')
    for p in axis.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        axis.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text
    return bp
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15,6))
create_subplot('loan_status',loan[loan['verification_status']=='Verified'],'Loan Status','Distribution of Verified Source Income on DTI','dti_cat','Set3',0, ax1)
create_subplot('loan_status',loan[loan['verification_status']=='Not Verified'],'Loan Status','Distribution of Not Verified Source Income on DTI','dti_cat','Set3',0, ax2)   
```
